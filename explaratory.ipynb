```python
# read by default 1st sheet of an excel file
population_df = pd.read_excel("/Users/amelchristy/Desktop/Python_Project/Real_Estate_Lehigh/uscities_population_details.xlsx")
```

```python
#remove the space on the zipcode 
population_df.head()
```

```python
# split the zip 
pop_df= population_df.assign(zips=population_df['zips'].str.split()).explode('zips')

pop_df.reset_index(drop=True, inplace=True)


pop_df.shape
```

```python
pop_df.head()
```

```python
print(pop_df.shape)
print(house_df.shape)
```

```python
# Convert the unique values in both columns to sets
unique_zips_population = set(pop_df['zips'].unique())
unique_zips_house = set(house_df['zipcode'].unique())

# Find the intersection of both sets to see which zip codes are present in both dataframes
common_zips = unique_zips_population.intersection(unique_zips_house)

# Print out the common zip codes
print(common_zips)

# Check the length of the common zip codes set to see how many matches there are
print(f"Number of matching zip codes: {len(common_zips)}")
```

```python
# Perforfm the inner join on 'zips' and 'zipcode' columns join house and population

joined1= pd.merge(house_df, pop_df, left_on='zipcode', right_on='zips', how='inner')
```

```python
joined1.shape
```

```python
# Drop the 'zipcode' column from the joined dataframe, as it's redundant
joined1 = joined1.drop(columns=['zips'])

# Drop all duplicate rows in the dataframe
joined1 = joined1.drop_duplicates()

```
